As expected the optimal VRP distance is shorter than both the nearest neighbor and savings heuristics. In terms of ordering the solutions by distances: Nearest Neighbor > Savings Heuristic > optimal solution. As a preface: it is difficult to find the optimal solution. Even with only 1 vehicle and 10 nodes there could be 3.6 million potential solutions, and in the above example there are clearly more than 10 nodes. 

The nearest neighbor heuristic initialises at the depot and appends the nearst feasible node to the end of the current route, in this case it'd represent a tour. If no feasible node is found for the current route, then the current route gets closed and we return to the depot to create a new route, where these 2 steps will be repeated until we finally visit every node. This heuristic is the least efficient out of the 3 methods presented here because of its short-sighted approach to finding nearby nodes. By only considering the closest nodes and giving up when that node can't be found instead of delving further potentially goods routes are rejected and then the heuristic forces a restart. 

The savings heuristic starts with the smallest cycles and keeps merging routes with the largest saving in cost in order to find the most efficient routes (tours), while checking **all** possible/feasible route merges. These steps are repeated until no more merges can be performed and we are left with the shortest routes. This method is notably less 'short sighted' than nearest neighbor and while it's technically a more expensive method of routing, it produced a distance that's only slightly above the optimal solution.
